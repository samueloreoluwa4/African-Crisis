### African crisis data analysis

```python
import pandas as pd
from pandas import DataFrame as df
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
crises = pd.read_csv('african_crises.csv')
crises.head()
```

```python
d = crises.groupby(crises['country'])
```

```python
dd = d.describe()
dd
```

```python
#Representation of inflation annual cpi with respect to the year and country
dci = crises[['country','year','inflation_annual_cpi']]
dcis = dci.set_index('country')
dcis.head()
```

```python
c_unique = crises.country.unique()
c_un = df(c_unique)
c_un
```

```python
# Plot of inflation annual cpi according to the country
for i in c_unique:
    dcis.loc[i].plot('year', 'inflation_annual_cpi', title= i)
    plt.show()
```

From the graph of inflation_annual_cpi to year above, the rate of inflation of annual cpi is extreme in the country of angola.
In the country of zimbabwe, the inflation rate is very small and can be consider to be negligible when compared to other states

### FOCUSING ON NIGERIA 
Analysing the data from nigeria

```python
dataN = crises.set_index("country").loc['Nigeria']
dataN.head()
```

```python
dataN.banking_crisis.value_counts()
#There was crisis for 11 years and no crisis for 49 years
```
